### Two options for subplots


```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10,5))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10, 5))

# plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

###  which one should you use (pyplot vs matplotlib OO method)
* when plotting something quickly, use first
* when plotting something more advanced, use second

```python
# plot for patients over age 50
over_50 = heart_disease[heart_disease.age > 50]
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter", x='age', y='chol', c='target')
```

```python
## OO method mixed with pyplot method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind="scatter", x='age', y='chol', c='target', ax=ax)
# ax.set_xlim([45, 100])
```

```python
# lets make it from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target);

# Customize the plot
ax.set(title="Heart disease and cholestrol levels",
      xlabel='Age', ylabel='Cholestrol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# lets add a horizontal line across x axis for mean value
ax.axhline(over_50.chol.mean(), linestyle='--')
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True)

# lets add data to ax0
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target) # this changes the color scheme

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        ylabel="Cholestrol")

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(over_50.chol.mean(), linestyle='--');

# lets add data to ax1
scatter = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target)

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(over_50.thalach.mean(), linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

## Customizing the Plots

```python
# Styles
plt.style.available
```

```python
car_sales.Price.astype(int).plot(); # default style
```

```python
# plt.style.use('dark_background')
plt.style.use('seaborn-v0_8')
car_sales.Price.astype(int).plot();
```

```python
# Create some data
x = np.random.randn(10, 4)
df = pd.DataFrame(x, columns=['a', 'b', 'c', 'd'])
ax = df.plot(kind='bar')
```

```python
# Customize the plot with set() method
ax = df.plot(kind='bar')
# add some title
ax.set(title='Random bargraph for dataframe', xlabel="Row number", ylabel="random")
# make the legend visible
ax.legend().set_visible(True)
```

```python
# plt.style.set('seaborn') can be used to set style 
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True)

# lets add data to ax0
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target,
                     cmap="winter") # this changes the color scheme

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        ylabel="Cholestrol")
# change the x limit
ax0.set_xlim([50, 80])

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(over_50.chol.mean(), linestyle='--');

# lets add data to ax1
scatter = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target,
                     cmap="summer")

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
# change the ax1 x axis limit
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(over_50.thalach.mean(), linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
